```python
import numpy as np
import matplotlib.pyplot as plt
## setup parameters and state variables
T = 50 # total time to simulate (msec)
dt = 0.125 # simulation time step (msec)
time = np.arange(0, T+dt, dt) # time array
t_rest = 0 # initial refractory time
## LIF properties
Vm = np.zeros(len(time)) # potential (V) trace over time
Rm = 1 # resistance (kOhm)
Cm = 10 # capacitance (uF)
tau_m = Rm*Cm # time constant (msec)
tau_ref = 4 # refractory period (msec)
Vth = 1 # spike threshold (V)
V_spike = 0.5 # spike delta (V)
## Input stimulus
I = 1.5 # input current (A)

## iterate over each time step
for i, t in enumerate(time):
    if t > t_rest:
        Vm[i] = Vm[i-1] + (-Vm[i-1] + I*Rm) / tau_m * dt
    if Vm[i] >= Vth:
        Vm[i] += V_spike
        t_rest = t + tau_ref
    
# plot membrane potential trace
plt.plot(time, Vm)
plt.title('Leaky Integrate-and-Fire Example')
plt.ylabel('U')
plt.xlabel('Time (msec)')
plt.ylim([0,2])
plt.show()
```

```python
IRange = []
len(time)
for i in range(len(time)):
    IRange.append(I)
plt.plot(time, IRange)
plt.title('I and Time')
plt.ylabel('I')
plt.xlabel('Time (msec)')
plt.show()
```

```python
t_rest = 0 # initial refractory time
## LIF properties
Rm = 1 # resistance (kOhm)
Cm = 10 # capacitance (uF)
tau_m = Rm*Cm # time constant (msec)
tau_ref = 4 # refractory period (msec)
Vth = 1 # spike threshold (V)
V_spike = 0.5 # spike delta (V)
Vm = np.zeros(len(time)) # potential (V) trace over time
f = []
i_s = []
for j in np.arange(1, 3, 0.1):
    counter = 0
    for i, t in enumerate(time):
        if t > t_rest:
            Vm[i] = Vm[i-1] + (-Vm[i-1] + j*Rm) / tau_m * dt
        if Vm[i] >= Vth:
            Vm[i] += V_spike
            t_rest = t + tau_ref
            if(counter == 1):
                f.append(1/t)
                i_s.append(j)
            counter += 1

#frequancy plot
plt.plot(i_s, f)
plt.title('frequency')
plt.ylabel('F')
plt.xlabel('I')
plt.show()
```

```python
import random

t_rest = 0 # initial refractory time
## LIF properties
randomVm = np.zeros(len(time)) # potential (V) trace over time
Rm = 1 # resistance (kOhm)
Cm = 10 # capacitance (uF)
tau_m = Rm*Cm # time constant (msec)
tau_ref = 4 # refractory period (msec)
Vth = 1 # spike threshold (V)
V_spike = 0.5 # spike delta (V)
## Input stimulus
I = 1.5 # input current (A)
randomIRange = []

for i, t in enumerate(time):
    tempRandI = random.randrange(1,5)
    randomIRange.append(tempRandI)
    if t > t_rest:
        randomVm[i] = randomVm[i-1] + (-randomVm[i-1] + tempRandI*Rm) / tau_m * dt
    if randomVm[i] >= Vth:
        randomVm[i] += V_spike
        t_rest = t + tau_ref  
plt.plot(time, randomVm)
plt.title('Leaky Integrate-and-Fire Example')
plt.ylabel('U')
plt.xlabel('Time (msec)')  
plt.show()
```
