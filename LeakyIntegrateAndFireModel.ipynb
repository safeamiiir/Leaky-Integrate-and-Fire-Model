```julia
using Plots
```

```julia
Tao_m = .030; #s
El = -.065; #V
Vreset = El;
Vth = -.050; #mV
Rm = 90; #MOmega
numSpikes = 0;
T = .2;
deltat = .00001;
t_arr = 0:deltat:T;
V = Vreset;
Ie_arr = 0:.00001:.002; #nA
r_isi = Float64[];
```

```julia
for Ie in Ie_arr
    numSpikes = 0;
    V = Vreset;
    for t in t_arr
        V = V + (deltat / Tao_m) * (-V + El + Rm * Ie);
        if V >= Vth
            V = Vreset;
            numSpikes = numSpikes + 1;
        end
    end
    if numSpikes > 0
        push!(r_isi, numSpikes / T)
    else
        push!(r_isi, 0)
    end

end
```

```julia
plotly()
plot(Ie_arr, r_isi, xlabel = "Ie (nA)",
                    ylabel = "r_isi (Hz)",
                    title = "Leaky Integrate and Fire Model")
```
